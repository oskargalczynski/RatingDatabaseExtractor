```python
import psycopg2


def top10movies(): #Using sql querry prints top 10 movies from database
    postgreSQL_select_Query = 'SELECT MOVIES.TITLE, AVG(RATINGS.RATING) FROM MOVIES, RATINGS ' \
                              'WHERE MOVIES.MOVIEID=RATINGS.MOVIEID ' \
                              'GROUP BY MOVIES.TITLE ORDER BY AVG(RATINGS.RATING) DESC LIMIT 10'

    cursor.execute(postgreSQL_select_Query)
    mobile_records = cursor.fetchall()

    print("Top 10 movies:")
    for row in enumerate(mobile_records):
        print(row[0] + 1, ".", row[1][0], ", avg rating = ", row[1][1])
    print()


def countMovies(): #Using sql querry prints number of all movies from database
    postgreSQL_select_Query = "SELECT * FROM MOVIES"

    cursor.execute(postgreSQL_select_Query)
    mobile_records = cursor.fetchall()

    print(f"There are {len(mobile_records)} movies")
    print()


def mostCommonGenre(): #Using sql querry prints most commor movie genre from database
    postgreSQL_select_Query = 'SELECT GENRES, COUNT(GENRES) AS VALUE_OCCURRENCE ' \
                              'FROM MOVIES ' \
                              'GROUP BY GENRES ' \
                              'ORDER BY VALUE_OCCURRENCE DESC ' \
                              'LIMIT 1'

    cursor.execute(postgreSQL_select_Query)
    mobile_records = cursor.fetchall()

    print(f"Most common genre is {mobile_records[0][0]}, {mobile_records[0][1]} occurences")
    print()


def mostRatingUser(): #Using sql querry prints 5 users with highest number of ratings
    postgreSQL_select_Query = 'SELECT USERID, COUNT(USERID) AS VALUE_OCCURRENCE ' \
                              'FROM RATINGS ' \
                              'GROUP BY USERID ' \
                              'ORDER BY VALUE_OCCURRENCE DESC ' \
                              'LIMIT 5'

    cursor.execute(postgreSQL_select_Query)
    mobile_records = cursor.fetchall()

    print("Top 5 rating users:")
    for row in enumerate(mobile_records):
        print(row[0] + 1, ". userId = ", row[1][0], ", ratings = ", row[1][1])
    print()


def firstAndLastRate(): #Using two sql querries prints both first and last rate from database
    postgreSQL_select_Query = 'SELECT TO_TIMESTAMP(TIMESTAMP), TITLE ' \
                              'FROM RATINGS, MOVIES ' \
                              'WHERE RATINGS.MOVIEID=MOVIES.MOVIEID AND ' \
                              'TIMESTAMP=(SELECT MIN(TIMESTAMP) FROM RATINGS) ' \
                              'LIMIT 1'

    cursor.execute(postgreSQL_select_Query)
    mobile_records = cursor.fetchall()

    print(f"First rate was: {mobile_records[0][0]}, on a movie {mobile_records[0][1]}")
    print()
    postgreSQL_select_Query = 'SELECT TO_TIMESTAMP(TIMESTAMP), TITLE ' \
                              'FROM RATINGS, MOVIES ' \
                              'WHERE RATINGS.MOVIEID=MOVIES.MOVIEID AND ' \
                              'TIMESTAMP=(SELECT MAX(TIMESTAMP) FROM RATINGS) ' \
                              'LIMIT 1'

    cursor.execute(postgreSQL_select_Query)
    mobile_records = cursor.fetchall()

    print(f"Last rate was: {mobile_records[0][0]}, on a movie {mobile_records[0][1]}")
    print()


def moviesFromYear(y): #Using sql querry prints movies from inserted year
    postgreSQL_select_Query = 'SELECT TITLE FROM MOVIES ' \
                              'WHERE TITLE LIKE \'%(' + str(y) + ')\''

    cursor.execute(postgreSQL_select_Query)
    mobile_records = cursor.fetchall()

    print(f"Movies from {y}: ")
    for row in enumerate(mobile_records):
        print(row[0] + 1, ".", row[1][0])
    print()


try:
    #Connects to sql database with credentials provided in docker-compose.yml file
    connection = psycopg2.connect(user="user",
                                  password="pass123",
                                  host="172.20.0.5",
                                  port="5432",
                                  database="mydb")
    cursor = connection.cursor() #Establishing connection
    #Exectuting all above functions
    countMovies()
    mostCommonGenre()
    top10movies()
    mostRatingUser()
    firstAndLastRate()
    moviesFromYear(1990)

except (Exception, psycopg2.Error) as error:
    print("Error while fetching data from database", error) #Message in case of connection error

finally:
    # closing database connection.
    if connection:
        cursor.close()
        connection.close()
        print("Connection is closed")
```
